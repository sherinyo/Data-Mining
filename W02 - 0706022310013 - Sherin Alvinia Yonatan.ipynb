```python
# 1. Load the dataset into Python (use Pandas)

import pandas as pd

url = "https://raw.githubusercontent.com/NathaliaMinoque/datasets/refs/heads/main/jabodetabek_house_price.csv"
df = pd.read_csv(url, sep=',')

# Saya menggunakan seperation ',' menyesuaikan dengan data di github yang menggunakan tanda koma sebagai pemisah

# 2. Inspect the dataset: Display the first 5 rows.

df.head()
```

```python
# KEY FINDINGS

# 1. Dari jumlah "count", dapat diketahui bahwa kolom "building_age" dan "year_built" memiliki banyak data yang kosong dengan total terisi hanya 2108 dari 3553 data;
# Di sisi lain, kolom seperti land_size_m2, building_size_m2, carports, maid_bathrooms, dan maid_bedrooms memiliki data yang kosong tetapi sedikit.

# 2. Dilihat dari Mean, Median, dan Mode untuk melihat distribusinya, kolom price_in_rp menunjukkan Positively Skewed Data dimana Median < Mean dan nilai antara max dan 75% sangat besar jaraknya.
# hal ini juga menunjukkan bahwa yang menjadi outliers adalah beberapa proerti yang sangat mahal.

# 3. Bedroom. bathroom, land_size_m2, dan building_size_m2 memiliki max value yang terlalu tinggi perbandingannya yang dapat menyebabkan potensi data error ketika menentukan mean dan median

# 4. Rata-rata building_age adalah 3.8an dengan median 1 tahun yang menunjukkan bahwa bahwa dataset ini memiliki data rumah yang barusan dibuat.
```

```python
# 3. Handle Missing Values: Identify missing values & decide and apply one strategy

# A. CEK PERSENTASE
# Calculate null values and percentage
null_counts = df.isnull().sum().sort_values(ascending=True)
null_percentages = (null_counts / len(df)) * 100

# Create a DataFrame for columns with null values
null_df = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentages})

# Filter for columns with null values
null_df = null_df[null_df['Null Count'] > 0]

# Format the percentage with two decimal places and % symbol
null_df['Null Percentage'] = null_df['Null Percentage'].map('{:.2f}%'.format)

null_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the horizontal bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.barplot(x='Null Count', y=null_df.index, data=null_df, orient='h', palette='rainbow')
plt.xlabel('Number of Null Values', fontsize=12)
plt.ylabel('Columns', fontsize=12)
plt.title('Null Values per Column', fontsize=14)

# Add labels to the bars
for i, v in enumerate(null_df['Null Count']):
    plt.text(v + 1, i, str(v), color='black', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Identify missing values
# B. MENGHAPUS KOLOM

# Pertimbangan Drop Kolom:
# - Karena didapatkan banyak jarak antara persentase data yang kosong, maka saya kategorikan (0-5%, 6%-30%, dan >30%) 
# dan menghapus secara manual.
# Yang manual dihapus karena pertimbangan untuk kedepannya visualisasi juga tidak mungkin dilakukan menggunakan Title, ads_id, url, address
# Nanti visualisasinya bisa dilakukan dengan mengelompokkan yang lebih penting spt disctrict, lat, long, dan lainnya. 

# Kelompokkan kolom
low_missing = null_percentages[(null_percentages > 0) & (null_percentages < 5)].index
moderate_missing = null_percentages[(null_percentages >= 5) & (null_percentages <= 30)].index
high_missing = null_percentages[(null_percentages > 30)].index

print("Low missing (<5%):", list(low_missing))
print("Moderate missing (5â€“30%):", list(moderate_missing))
print("High missing (>30%):", list(high_missing))

```

```python
# Penerapan

# a. Low missing (0-5%) -> isi median (numeric) atau mode (categorical)
for col in low_missing:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)

# b. Moderate missing (6-30%) -> median (numeric) atau mode (categorical)
for col in moderate_missing:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)

# c. High missing (> 30%) -> drop kolom
df.drop(columns=high_missing, inplace=True)

# d. Manual Drop
# Kenapa ini yang didrop? ads_id dan url tidak ada hal yang mau divisualisasikan nantinya dan karena saya menggunakan lat long, 
# maka tidak memerlukan lagi menggunakan district.
manual_drop = ['ads_id', 'url', 'district']  
dropped_manual = [col for col in manual_drop if col in df.columns]
df.drop(columns=dropped_manual, inplace=True, errors='ignore')

# Cek
print("Cek Missing Values:")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# Berdasarkan boxplot, ditemukan banyak outlier pada beberapa kolom, akan dilakukan pengurangan jumlah data yang diambil 
# untuk menyeimbangkan data

# Saya menggunakan angka ini, untuk menyesuaikan dengan kriteria rumah yang realistis
# Hapus outlier ekstrem kategori numeric
df = df[
    (df['bedrooms'] <= 10) &  
    (df['land_size_m2'] <= 300) &
    (df['building_size_m2'] <= 300)
]

# Boxplot untuk Price in Rp
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price_in_rp'])
plt.title('Boxplot of price_in_rp')
plt.xlabel('Jumlah')
plt.show()

# Boxplot untuk land_size_m2
plt.figure(figsize=(12, 8))
sns.boxplot(x=df['land_size_m2'])
plt.title('Boxplot of land_size_m2')
plt.xlabel('Jumlah')
plt.show()

# Boxplot untuk building_size_m2
plt.figure(figsize=(12, 8))
sns.boxplot(x=df['building_size_m2'])
plt.title('Boxplot of building_size_m2')
plt.xlabel('Jumlah')
plt.show()

# Boxplot untuk bedrooms
plt.figure(figsize=(12, 8))
sns.boxplot(x=df['bedrooms'])
plt.title('Boxplot of bedrooms')
plt.xlabel('Jumlah')
plt.show()
```

```python
# 5. Feature engineering (binning/new feature): Create new feature using binning (if any).

# Saya menggunakan 3 kolom untuk dibuat bin untuk mempermudah nanti visualisasi berdasarkan kategori
# Contoh: luas bangunan berpengaruh dengan harga rumah maka dari itu perlu kita kategorikan untuk memudahkan visualisasi.

# price_in_rp
bins_price = [0, 1e9, 3e9, 5e9, df['price_in_rp'].max()]
labels_price = ['Murah', 'Sedang', 'Mahal', 'Sangat Mahal']

df['price_category'] = pd.cut(df['price_in_rp'], bins=bins_price, labels=labels_price)

# cek
df[['price_in_rp', 'price_category']].head(10)
```

```python
# land_size_m2

bins_landsize = [0, 100, 150, df['land_size_m2'].max()]
labels_landsize = ['Kecil', 'Sedang', 'Besar']

df['land_size_category'] = pd.cut(df['land_size_m2'], bins=bins_landsize, labels=labels_landsize, duplicates='drop')
df[['land_size_m2', 'land_size_category']].head(10)
```

```python
#building_size_m2

bins_buildingSize = [0, 100, 150, df['building_size_m2'].max()]
labels_buildingSize = ['Kecil', 'Sedang', 'Besar']

df['building_size_category'] = pd.cut(
    df['building_size_m2'], bins=bins_buildingSize, labels=labels_buildingSize, duplicates='drop'
)

df[['building_size_m2', 'building_size_category']].head(10)
```
